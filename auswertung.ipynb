```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df_1 = pd.read_csv('Data/ingredients.csv')
df_2 = pd.read_csv('Data/recipes.csv')
```

```python
#Azahl und unterschiedliche Werte der Ids und der Zutaten
ingredients = pd.unique(df_1['ingredients'])
len_ingredients = len(pd.unique(df_1['ingredients']))
len_cuisine = len(pd.unique(df_2['ID']))

#Dataframe 40k x 6k, danach id und cuisine hinzufügen
df_3 = pd.DataFrame(np.zeros((len_cuisine, len_ingredients)), columns=ingredients)
df_3.insert(0, 'id', df_2['ID'])
df_3.insert(1, 'cuisine', df_2['cuisine'])
df_3
```

```python
#Für die nächsten 50k einträge den wert in der while schleife erhöhen und i=0 rausnehmen
i=0
while i <= 50000:
    #ID und Zutat des iten eintrags
    id_selected = df_1.loc[[i]]['ID_recipe'].values[0]
    ingredient_selected = df_1.loc[[i]]['ingredients'].values[0]

    #Diese zeile funktioniert gerade nicht, sie soll die zelle reihe nach id, spalte nach zutat auf 1 setzen. Ich komme gerade nicht an meine
    #anderen noteboos ran, erkennst du den Fehler?
    df_3.loc[df_3['id'] == id_selected][ingredient_selected] = 1

    i=i+1
```
